```python
import numpy as np

# Simple activation functions and their derivatives
def sigma(x):
    return np.tanh(x)

def sigma_prime(x):
    return 1 - np.tanh(x)**2

# Simulation of fractional Brownian motion increments
def simulate_fbm_increments(H, size, L):
    n, m = size
    increments = np.zeros((L, n, m))

    # Generate increments independently for each weight
    for i in range(n):
        for j in range(m):
            cov = np.zeros((L+1, L+1))
            for s in range(L+1):
                for t in range(L+1):
                    cov[s, t] = 0.5 * (abs(s)**(2*H) + abs(t)**(2*H) - abs(s - t)**(2*H))

            chol_cov = np.linalg.cholesky(cov + 1e-10*np.eye(L+1))
            fbm_samples = chol_cov @ np.random.randn(L+1)
            fbm_increments = np.diff(fbm_samples) / (L**H)
            increments[:, i, j] = fbm_increments

    return increments


# First ResNet version (with fractional Brownian increments)
def resnet_standard(X0, H, L, beta):
    X = X0
    W_fbm = simulate_fbm_increments(H, (n, n), L)
    scaling = 1/(L**beta)
    for k in range(L):
        X = X + scaling*W_fbm[k] @ sigma(X)
    return X

# Second ResNet version with fractional Brownian increments and correction term
def resnet_fbm(X0, H, C_H, L, beta):
    X = X0
    W_fbm = simulate_fbm_increments(H, (n, n), L)
    scaling = 1/(L**beta)
    for k in range(L):
        correction_term = (C_H / L**(2*beta)) * sigma_prime(X) * sigma(X)
        X = X + scaling*W_fbm[k] @ sigma(X) + correction_term

    return X

# Example Usage
if __name__ == '__main__':
    np.random.seed(42)

    # Parameters
    n = 10  # Dimension
    L = 100  # Layers
    H = 0.75  # Hurst index
    C_H = 0.5  # Constant for correction term

    # Initial input
    X0 = np.random.randn(n)

    # Run standard ResNet with fractional Brownian increments
    output_standard = resnet_standard(X0, H, L,0.5)

    # Run ResNet with fractional Brownian motion increments and correction
    output_fbm = resnet_fbm(X0, H, C_H, L,0.5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set grid for H and beta
H_vals = np.linspace(0., 1.0, 10)
beta_vals = np.linspace(0., 1.0, 10)

# Parameters
n = 10
L = 100
C_H = 0.5
X0 = np.random.randn(n)
h0 = np.linalg.norm(X0)

# Prepare result matrix
log_change = np.zeros((len(H_vals), len(beta_vals)))

# Compute hL and log((hL - h0)/h0)
for i, H in enumerate(H_vals):
    for j, beta in enumerate(beta_vals):
        output = resnet_fbm(X0, H, C_H, L, beta)
        hL = np.linalg.norm(output)
        log_change[i, j] = np.log((hL - h0) / h0)

# Plotting
plt.figure(figsize=(10, 6))
sns.heatmap(log_change, xticklabels=np.round(beta_vals, 2), yticklabels=np.round(H_vals, 2),
            cmap='coolwarm', cbar_kws={'label': r'$\ln\left(\frac{h_L - h_0}{h_0}\right)$'})
plt.xlabel(r'$\beta$')
plt.ylabel(r'$H$')
plt.title(r'Log relative change of $\|X_L\|$ over grid of $\beta$ and $H$')
plt.tight_layout()
plt.show()
```
